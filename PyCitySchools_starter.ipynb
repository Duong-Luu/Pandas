```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
data_school = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, data_school, how="left", on=["school_name", "school_name"])

school_data_complete
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

```python
school = school_data_complete["school_name"].unique()
number_school = len(school)
number_school
```

```python
student = school_data_complete["student_name"].count()
student
```

```python
budget = school_data_complete["budget"].unique()
budget_total = budget.sum()
```

```python
math = school_data_complete["math_score"].mean()
math
```

```python
read = school_data_complete["reading_score"].mean()
read
```

```python
passing_rate = (math + read)/2
passing_rate
```

```python
math_70 = school_data_complete.loc[school_data_complete['math_score'] >= 70]['math_score'].count()
math_total = school_data_complete["math_score"].count()
math_per = (math_70 / math_total) * 100
math_per
#a = total / math_count
#print(a)
```

```python
read_70 = school_data_complete.loc[school_data_complete['reading_score'] >= 70]['reading_score'].count()
read_total = school_data_complete['reading_score'].count()
read_per = (read_70 / read_total) * 100
read_per
```

```python
df = pd.DataFrame({
    'Total School' : [number_school],
    'Total Students' : [student],
    'Total Budget' : [budget_total],
    'Average Math Score' : [math],
    'Average Reading Score' : [read],
    '% Passing Math' : [math_per],
    '% Passing Reading' : [read_per],
    '% Overall Passing Rate' : [passing_rate]
    
})

df["Total Students"] = df['Total Students'].map("{:,}".format)
df["Total Budget"] = df['Total Budget'].map("${:,.2f}".format)


df
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
school_data = pd.merge(student_data, data_school, on="school_name")
school_data
```

```python
group_school = school_data.groupby('school_name')
```

```python
sch_types = data_school.set_index('school_name')['type']

sch_budget = data_school.set_index('school_name')['budget']
```

```python
group_school.count()
```

```python
total_stu = group_school['student_name'].count()
total_stu
```

```python
#stu_budget = school_data.set_index('school_name')['budget'] / school_data.set_index('school_name')['size']

stu_budget = sch_budget / data_school.set_index('school_name')['size']

stu_budget
```

```python
math_ave = group_school['math_score'].mean()
read_ave =  group_school['reading_score'].mean()
```

```python
school_data[school_data['math_score'] >= 70].count()
```

```python
passing_math_percent = 100*(school_data[school_data['math_score'] >= 70].groupby('school_name')['student_name'].count() / group_school['student_name'].count())
passing_read_percent = 100*(school_data[school_data['reading_score'] >= 70].groupby('school_name')['student_name'].count() / group_school['student_name'].count())

overall_passing_per = 100*(school_data[(school_data['math_score'] >= 70) & (school_data['reading_score'] >= 70)].groupby('school_name')['student_name'].count() / group_school['student_name'].count())
```

```python
school_summary = pd.DataFrame({
    'School Type' : sch_types,
    'Total Student' : total_stu,
    'Total School Budget' : sch_budget,
    'Per Student Budget'  : stu_budget,
    'Average Math Score'  : math_ave,
    'Average Reading Score' : read_ave,
    '% Passing Math'  : passing_math_percent,
    '% Passing Reading' : passing_read_percent,
    '% Overall Passing Rate' : overall_passing_per
})


school_summary["Total School Budget"] = school_summary['Total School Budget'].map("${:,.2f}".format)
school_summary["Per Student Budget"] = school_summary['Per Student Budget'].map("${:,.2f}".format)
school_summary["Average Math Score"] = school_summary['Average Math Score'].map("{:.6f}".format)
school_summary["Average Reading Score"] = school_summary['Average Reading Score'].map("{:.6f}".format)
school_summary["% Overall Passing Rate"] = school_summary['% Overall Passing Rate'].map("{:.6f}".format)



school_summary
```

```python
top_5 = school_summary.sort_values('% Overall Passing Rate', ascending = False).head(5)

top_5
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
school_data
```

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
budget_per_student = school_data['budget']/school_data['size']
```

```python
school_data['Spending Ranges (Per Student)'] = pd.cut(budget_per_student, spending_bins, labels = group_names)
school_data
```

```python
spending_group = school_data.groupby(['Spending Ranges (Per Student)'])
spending_group.mean()
```

```python
avg_math_score = spending_group['math_score'].mean()
avg_reading_score = spending_group['reading_score'].mean()

avg_math_score
```

```python
percent_of_math = 100* (school_data[school_data['math_score'] >= 70].groupby('Spending Ranges (Per Student)')['Student ID'].count() / spending_group['Student ID'].count())

percent_of_read = 100* (school_data[school_data['reading_score'] >= 70].groupby('Spending Ranges (Per Student)')['Student ID'].count() / spending_group['Student ID'].count())

percent_of_overall = 100* (school_data[(school_data['math_score'] >= 70) & (school_data['reading_score'] >= 70) ].groupby('Spending Ranges (Per Student)')['Student ID'].count() / spending_group['Student ID'].count())

```

```python
pd.DataFrame({
    'Spending Ranges (Per Student)' : group_names,
    'Average Math Score' : avg_math_score,
    'Average Reading Score' : avg_reading_score,
    '% Passing Math' : percent_of_math,
    '% Passing Reading' : percent_of_read,
    '% Passing Overall' : percent_of_overall
    
})
```
